##### 1. Создайте одномерный массив из чисел от 1 до 10. Затем преобразуйте его в двумерный массив 2x5. Найдите сумму всех элементов массива, а также среднее арифметическое.

```python
import numpy as np
```

```python
# Место для вашего кода :)
arr = np.arange(1, 11)
arr2 = arr.reshape(2, 5)

sum = np.sum(arr2)
mean = np.mean(arr2)
print(f"Двумерный массив 2x5:")
print(arr2)
print(f"Сумма всех элементов массива: {sum}")
print(f"Среднее арифметическое элементов массива: {mean}")
```

##### 3. Создайте массив из 20 случайных чисел. Найдите:
1. Минимальное и максимальное значения
2. Медиану
3. Стандартное отклонение

```python
# Место для вашего кода :)
arr = np.random.randint(1, 101, 20)
print(f"Минимальное значение {arr.min()}, а максимальное {arr.max()}")
print(f"Медиана - {np.median(arr)}")
print(f"Стандартное отклоненение - {round(np.std(arr), 4)}")
```

##### 4. Создайте массив случайных чисел от 0 до 100 (10 элементов). Определите, какие из них чётные. Замените все нечётные элементы на 0.

```python
# Место для вашего кода :)
arr = np.random.randint(1, 101, 20)
print(arr)
for i in range(len(arr)):
    if (arr[i] % 2) != 0:
        arr[i] = 0
print(arr)
```

##### 5.1 Создайте матрицу произвольного размера. 
1. Найдите транспонированную матрицу.
2. Найдите обратную матрицу, если это возможно.
3. Найдите собственные значения и собственные вектора.

```python
# Место для вашего кода :)
matrix = np.random.randint(0, 101,(np.random.randint(1, 11),np.random.randint(1, 11)))
print(matrix)
print("\nТранспонирвоанная матрица")
print(np.transpose(matrix))
if (matrix.shape[0] == matrix.shape[1]) and (np.linalg.det(matrix) != 0):
    print("\nОбратная матрица")
    print(np.linalg.inv(matrix))
else:
    print("\nМатрица вырождена")

if (matrix.shape[0] == matrix.shape[1]):
    print("Собственные значения:", np.linalg.eig(matrix)[0])
    print("Собственные векторы:\n", np.linalg.eig(matrix)[1])
```

##### 5.2 Создайте две матрицы 4x4, заполненные случайными числами. Выполните следующие операции:

1. Найдите их матричное произведение.
2. Объедините две матрицы в тензор/в одну матрицу 4x8.

```python
# Место для вашего кода :)
matrix = np.random.randint(0, 101,(4,4))
matrix2 = np.random.randint(0, 101,(4,4))
print("Матричное произведение:")
print(matrix @ matrix2)
print("Объединенная матрица:") #тензор через np.stack, но задание было или тензор или одну матрицу.
print(np.concatenate((matrix, matrix2), axis=1))
```

3. Отсортируйте каждый слой по возрастанию.

```python
# Место для вашего кода :)
arr = np.random.random((3, 3, 3))
print(arr)
print("\nСумма по оси 0:\n", np.sum(arr, axis=0))
print("\nСумма по оси 1:\n", np.sum(arr, axis=1))
print("\nСумма по оси 2:\n", np.sum(arr, axis=2))

print("\nМинимальные элементы каждого слоя:\n", np.min(arr, axis=(1, 2)))

print("\nТрехмерный массив после сортировки:\n", np.sort(arr, axis=1))
```
